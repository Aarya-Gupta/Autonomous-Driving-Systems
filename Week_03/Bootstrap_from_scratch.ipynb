```python
# I know that the distribution is uniform(-10, +10)
import random
import statistics
import matplotlib.pyplot as plt

def staticalMeasures (List):
    mean_sample_X = statistics.mean(List)
    # mode_sample_X = statistics.mode(List)
    # median_sample_X = statistics.median(List)
    std_sample_X = statistics.stdev(List)
    return mean_sample_X, std_sample_X

def generateProbDictionary(List) -> dict :
    dict_X = {}
    for i in List :
        if (i not in dict_X) :
            dict_X[i] = sample_X.count(i)/len(List)

    sorted_items = sorted(dict_X.items())
    # Create a new dictionary from the sorted list of key-value pairs
    sorted_dict = dict(sorted_items)
    return sorted_dict

def returnBootstrapSample (dictionary, size_of_list)-> list :
    # Create a list of numbers and their weights
    numbers = list(dictionary.keys())
    weights = list(dictionary.values())

    # Generate a list of 1000 numbers based on the probabilities
    result_list = random.choices(numbers, weights=weights, k=size_of_list)
    return result_list

def calculateMSE (D1, D2) :
    E_X2 = sum((x**2) * p for x, p in zip(D1.keys(), D1.values()))
    E_Y2 = sum((y**2) * p for y, p in zip(D2.keys(), D2.values()))
    E_XY = sum(x * y * p1 * p2 for x, y, p1, p2 in zip(D1.keys(), D2.keys(), D1.values(), D2.values()))

    squared_diffs = [(x - y) ** 2 for x, y in zip(D1.keys(), D2.keys())]
    MSE = sum(diff * list(D1.values())[i] * list(D2.values())[i] for i, diff in enumerate(squared_diffs))
    return MSE
    # print(MSE)
    # return (E_X2 - 2*E_XY + E_Y2)/len(D1)
    
def printObservation (sample_X) :
    mean_sample_X, std_sample_X = staticalMeasures (sample_X)
    print ("Mean : {}, Std. Deviation : {}".format(mean_sample_X, round(std_sample_X,2)))

def plot(num_prob, mse) :
    import matplotlib.pyplot as plt
    # Extract the numbers and probabilities into separate lists
    numbers = list(num_prob.keys())
    probabilities = list(num_prob.values())

    # Create a figure and axis objects
    fig, ax = plt.subplots()

    # Plot the data as a bar chart
    ax.bar(numbers, probabilities)

    # Set the title and axis labels
    ax.set_title("Sample vs Probability")
    ax.set_xlabel("Sample")
    ax.set_ylabel("Probability")
    if (mse!="NO NEED"):
        ax.axvline(x=mse, color='r', linestyle='--', linewidth=2, label="MSE")
    # Show the plot
    plt.show()

# --------------------------------Initial Conditions Started----------------------------------------
sample_X = []
for i in range (10000) :
    # Very Important Assumption is that I am taking random.randint()
    # Discretising a continous waveform.
    # Random.randint is a uniform distribution.
    # sample_X.append(random.randint(-10, +10));
    sample_X.append(round(random.gauss(0, 1), 1));
print ("Observed Samples : \n{}".format(sample_X))

mean_sample_X, std_sample_X = staticalMeasures (sample_X)
print ("Observed Mean : {}, Observed Std. Deviation : {}".format(mean_sample_X, round(std_sample_X,2)), end = "\n\n")
# --------------------------------Initial Conditions Ended-------------------------------------------

# Now, let's suppose that the sample_X has been randomly sampled, without the knowledge of the range -10 to +10;
# Or that the distribution is uniform
# Assuming a distribution on the basis of the observed samples.

dict_X = generateProbDictionary(sample_X)
print ("Given Information \n <PARENT DISTRIBUTION>:\n\n{}".format(dict_X))
print("The dictionary shows the sampled distribution of items along with the probability assigned to it.")

new_list = returnBootstrapSample(dict_X, 1000)
MSE_start = calculateMSE(dict_X, generateProbDictionary(new_list))
plot(dict_X, "NO NEED")
print("\n\n")

sample_X = new_list

# Generating bootstrap samples, and modifying the distribution; in order to minimise MSE.
count = 1
for i in range (0, 10) :
    # Creating a new distribution (which is known to us), on the basis of the values obtained; 
    # Assigning probability to each unique item in the randomly sampled list, on the basis of the occurence in that list.
    dict_X = generateProbDictionary(sample_X)
    # Bootstraping samples from the above dictionary.
    new_list = returnBootstrapSample(dict_X, 1000)
    # Updating the sample_X to the new_list, so that it can be used for generating emperical distribution for the next iteration.
    new_MSE = calculateMSE(dict_X, generateProbDictionary(new_list))
    
    # Minimising the MSE.
    if (new_MSE < MSE_start) :
        print ("Iteration No. {} \n{}".format(i+1, dict_X))
        print ("Now, the dictionary itself shows an emperical distribution of items along with the probability assigned to it.")
        print ("MSE : ", MSE_start)
        printObservation(new_list)
        print()
        plot(dict_X, MSE_start)
        MSE_start = new_MSE
        count+=1
    sample_X = new_list
```

Aim is to minimise the MSE, so that the predicted distribution is nearest to the actual distribution, from where the values are being derived.

As the no. of iterations is increasing, no. of items in the bootstrapped sample is decreasing; this is because of the function defination of returnBootstrapSample, in which randomly items are chosen for the new list on the basis of the probability associated with the current items; hence it is possible that some items with lesser probability might not be considered in the new list, hence removing their effect.
